```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
df = pd.read_csv('Swiz_clinc.csv')
df = df.dropna()
```

```python
df['geometry'] = [Point(xy) for xy in zip(df.Longitude, df.Latitude)]

gdf = gpd.GeoDataFrame(df, geometry='geometry')
swiss_regions = gpd.read_file("health_centers.shp")

swiss_regions.plot(column='Type', cmap='jet', legend=True, figsize=(15, 16),
                   legend_kwds={'fontsize': 'small'})

plt.title("Distribution of Health Centers across Different Regions in Switzerland Filtered by Type", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(df.Longitude, df.Latitude, c=df.Rating, cmap='jet')
plt.colorbar()
plt.title("Relationship between Health Center Location and Ratings", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(df.Longitude, df.Latitude, c=df.Reviews, cmap='jet')
plt.colorbar()
plt.title("Relationship between Health Center Location and Reviews", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(df['Reviews'], df['Rating'])
plt.title("Comparison of Health Centers in Terms of their Ratings and Reviews", fontsize = 15)
plt.xlabel("Number of Reviews", fontsize = 13)
plt.ylabel("Ratings", fontsize = 13)
plt.show()
```
