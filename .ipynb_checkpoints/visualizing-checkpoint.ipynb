```python
import mpl_toolkits
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('Swiz_clinc.csv')
```

```python
# Load the data into a GeoDataFrame
df = pd.DataFrame({'Names': ['Gesundheitszentrum Fricktal AG', 'Klinik Professor Sailer', 'Kantonsspital Baselland Liestal', 'Klinik Wyssh√∂lzli', 'Centre Jean-Violette'],
                   'Latitude': [47.508943, 47.369394, 47.488018, 47.184946, 46.194171],
                   'Longitude': [8.016782, 8.566445, 7.731157, 7.715649, 6.144164]})

# Create a geometry column with the latitude and longitude
df['geometry'] = gpd.points_from_xy(df.Longitude, df.Latitude)
gdf = gpd.GeoDataFrame(df, geometry='geometry')

# Plot the data on a map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='white', edgecolor='black', figsize=(15,10))
gdf.plot(ax=ax, markersize=100, color='red')

plt.show()
```

```python


# drop NaN values from the dataframe
df = df.dropna()

# plot the latitude and longitude with the ratings and reviews
plt.figure(figsize = (15, 7))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Rating'], s=df['Reviews'] * 29 , cmap='viridis')
plt.colorbar()

# set the x and y axis labels
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# show the plot
plt.show()
```
