```python
import requests
from time import sleep
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor
```

```python
# Initialize a counter for the page number
page_number = 1

# Initialize an empty list to store the clinic listings
clinic_listings = []
clinic_addresses = []
clinic_type = []
clinic_reviews = []
clinic_count_reviews = []
links = []

# Loop until there are no more pages
while True:
    # Construct the URL with the current page number
    url = f"https://www.local.ch/en/q/Switzerland/clinique?page={page_number}&rid=j6VE&slot=tel"

    # Send a request to the URL and get the response
    response = requests.get(url, timeout = 20)

    # Check if the response was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Find all div elements with class "lui-card"
        listings = soup.find_all("div", class_="card-info-element")
        
        addresses = soup.find_all('div', class_ = "card-info-address")
        
        cl_type = soup.find_all('div', class_ = "card-info-category")
        
        link = soup.find_all('div', "entry-card-info-element entry-card-info-element-first")

        # Check if any listings were found
        if listings:
            # Add the listings to the list of clinic listings
            clinic_listings.extend(listings)
            clinic_addresses.extend(addresses)
            links.extend(link)
            clinic_type.extend(cl_type)
            clinic_reviews.extend([requests.get(i.find('a')['href'],  timeout = 20).content for i in links if requests.get(i.find('a')['href'], 
                                                timeout = 20).status_code == 200])
            clinic_count_reviews.extend([BeautifulSoup(requests.get(i.find('a')['href'], timeout = 20).content, "html.parser").find_all('div', "feedback-summary-ratings-value lui-margin-bottom-xxs") 
                                                                    for i in links if requests.get(i.find('a')['href'], timeout = 20).status_code == 200])

            # for i in links:
            #     url2 = i.find('a')['href']
            #     r = requests.get(url2, timeout = 20)
            #     if r.status_code == 200:
            #         soup2 = BeautifulSoup(r.content, "html.parser")
            #         rev = soup2.find_all('div', "feedback-summary-ratings-value lui-margin-bottom-xxs")
            #         clinic_reviews.extend(rev)
            #         no_review = soup2.find_all('div', "text-muted text-nowrap lui-margin-top-xs")
            #         clinic_count_reviews.extend(no_review)
            # Increment the page number
            print(f'Page {page_number} completed')
            page_number += 1
        else:
            # If no listings were found, break the loop
            break
    else:
        # If the response was not successful, break the loop
        break

# Print the number of clinic listings found
print(f"Number of clinic listings: {len(clinic_listings)}")
```
